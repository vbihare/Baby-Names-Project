```python
def last_letter_counter(lst):
    return zip(*sorted(Counter(lst['name'].apply(lambda x: x[-1])).items()))
```

```python
male_name_count_labels,male_name_count_values = last_letter_counter(males_from_file)
female_name_count_labels,female_name_count_values = last_letter_counter(females_from_file)
```

```python
random_all_name = list(df_from_file['name'].unique())[:100]

name_1940_dic = {}
for name in random_all_name:
    temp = df_from_file[df_from_file['name'] == name].groupby(['year']).count()
    temp = temp.reset_index()
    var = np.var(list(temp[(temp['year'] >= 1940) & (temp['year'] < 1950)]['name']))
    name_1940_dic[name] = var
```

```python
name_1930_dic = {}
for name in random_all_name:
    temp = df_from_file[df_from_file['name'] == name].groupby(['year']).count()
    temp = temp.reset_index()
    var = np.var(list(temp[(temp['year'] >= 1930) & (temp['year'] < 1940)]['name']))
    name_1930_dic[name] = var
```

```python
name_1950_dic = {}
for name in random_all_name:
    temp = df_from_file[df_from_file['name'] == name].groupby(['year']).count()
    temp = temp.reset_index()
    var = np.var(list(temp[(temp['year'] >= 1950) & (temp['year'] < 1960)]['name']))
    name_1950_dic[name] = var
```

```python
variance1930 = pd.DataFrame.from_dict(name_1930_dic,orient='index')
variance1940 = pd.DataFrame.from_dict(name_1940_dic,orient='index')
variance1950 = pd.DataFrame.from_dict(name_1950_dic,orient='index')
```

```python
variance1930.reset_index(inplace=True)
variance1940.reset_index(inplace=True)
variance1950.reset_index(inplace=True)
variance1930.columns = ['Name','Variance']
variance1940.columns = ['Name','Variance']
variance1950.columns = ['Name','Variance']
```

```python
variance1930.sort_values('Variance',ascending=False,inplace=True)
variance1940.sort_values('Variance',ascending=False,inplace=True)
variance1950.sort_values('Variance',ascending=False,inplace=True)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation='vertical')
plt.bar('Name','Variance',data=variance1930)
plt.bar('Name','Variance',data=variance1940)
plt.bar('Name','Variance',data=variance1950)
plt.legend(['1930s','1940s','1950s'],loc='upper right',title='Years')
```
